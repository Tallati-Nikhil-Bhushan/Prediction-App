```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.ensemble import RandomForestClassifier
import pickle
```

```python
kidney_df = pd.read_csv('../Datasets/kidney/kidney_disease.csv')
kidney_df.head()
```

```python
kidney_df.columns
```

```python
kidney_df.info()
```

```python
kidney_df.describe()
```

```python
kidney_df.shape
```

```python
kidney_df=kidney_df.drop('id',axis=1)
kidney_df
```

```python
kidney_df.isnull().sum()
```

```python
kidney_df['classification'].replace("ckd\t","ckd",inplace=True)

kidney_df['dm'].replace(["\tno","\tyes"," yes"],["no","yes","yes"],inplace=True)

kidney_df['cad'].replace(["\tno"],["no"],inplace=True)

kidney_df['rc'].replace("\t?",np.nan, inplace=True)

kidney_df['wc'].replace("\t?",np.nan, inplace=True)

kidney_df['pcv'].replace(["\t?","\t43"],np.nan, inplace=True)
```

```python
NumericalColumns = ['age','bp','al','su','bgr','bu','sc','sod','pot','hemo']
CategoricalColumns = ['rbc','sg','pc','pcc','ba','pcv','wc','rc','htn','dm','cad','appet','pe','ane']

for columnName in CategoricalColumns:
    kidney_df[columnName].fillna(kidney_df[columnName].mode()[0],inplace=True)

for columnName in NumericalColumns:
    kidney_df[columnName].fillna(kidney_df[columnName].mean(), inplace=True)
```

```python
kidney_df.isnull().sum()
```

```python
encodeColumn = ['rbc','pc' ,'pcc' ,'ba' ,'htn' ,'dm' ,'cad' ,'appet' ,'pe' ,'ane']
kidney_df = pd.get_dummies(kidney_df , columns=encodeColumn , prefix=encodeColumn , drop_first=True)
```

```python
kidney_df['classification'].replace(["ckd","notckd"],[1,0], inplace=True)
```

```python
kidney_df.head()
```

```python
kidney_df['classification'].value_counts()
```

```python
kidney_df.groupby('classification').mean()
```

```python
X = kidney_df.drop(columns='classification',axis=1)
Y = kidney_df['classification']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=100)
```

```python
rfc = RandomForestClassifier()
rfc.fit(X_train, Y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
print(confusion_matrix(Y_test,rfc_pred))
```

```python
print(classification_report(Y_test,rfc_pred))
```

```python
filename='../prediction_models/kidney_model.sav'
pickle.dump(rfc,open(filename,'wb'))
```

```python
loaded_model = pickle.load(open(filename,'rb'))
```

```python
input_data = np.asarray([13,68.0,70.0,1.02,1.0169491525423728,0.45014245014245013,98.0,86.0,4.6,135.0,3.4,9.8,41,9800,5.2,1,1,0,0,1,1,1,1,1])

input_data_reshaped = input_data.reshape(1,-1)

prediction = loaded_model.predict(input_data_reshaped)

if (prediction[0] == 1):
  print('Having Kidney Disease')
else:
  print('Not Having Kidney Disease')
```
